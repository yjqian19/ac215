# Loading the Unsplash Research dataset in Pandas dataframes

This notebooks is an example of how to load the Unsplash Research dataset in Pandas dataframes for analysis.


## Loading libraries

```python
import numpy as np
import pandas as pd
import glob
```

## Loading the datasets in Pandas

Make sure that you correctly point to the correct path.

```python
path = './'
documents = ['photos', 'keywords', 'collections', 'conversions', 'colors']
datasets = {}

for doc in documents:
  files = glob.glob(path + doc + ".csv000")

  subsets = []
  for filename in files:
    df = pd.read_csv(filename, sep='\t', header=0)
    subsets.append(df)

  datasets[doc] = pd.concat(subsets, axis=0, ignore_index=True)
```

## Exploring the datasets

Here are the first couple of rows from each dataset, as an example.

Enjoy the exploration!

```python
datasets['photos'].head()
```

```python
photos = datasets['photos']
photos.filter(like='exif_')
```

```python
import matplotlib.pyplot as plt
# count by photographer_username
user_counts = photos['photographer_username'].value_counts().head(20)

# draw bar chart
plt.figure(figsize=(10,6))
user_counts.plot(kind='bar')

plt.title("Top 10 Photographers by Photo Count")
plt.xlabel("Photographer Username")
plt.ylabel("Number of Photos")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
datasets['keywords'].head()
```

```python
keywords = datasets['keywords']
# count by keyword
user_counts = keywords['keyword'].value_counts().head(40)

# draw bar chart
plt.figure(figsize=(10,6))
user_counts.plot(kind='bar')

plt.title("Top 10 Keywords by Photo Count")
plt.xlabel("Keyword")
plt.ylabel("Number of Photos")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
